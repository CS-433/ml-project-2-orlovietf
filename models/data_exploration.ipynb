```python
import pandas as pd

df = pd.read_csv("../train_updated.csv")

sequences = df["protein_sequence"].tolist()
tm = df["tm"].values
ph = df["pH"].values
data_sources = df["data_source"].values
```

```python
df = pd.read_csv("../test.csv")

test_sequences = df["protein_sequence"].tolist()
```

```python
import matplotlib.pyplot as plt

sequence_lengths = [len(sequence) for sequence in sequences]
plt.hist(sequence_lengths, bins=100, edgecolor='black')

# Increase font sizes
plt.title('Training Sequence Lengths', fontsize=20)
plt.xlim(0, 3000)
plt.xlabel('Sequence Lengths', fontsize=18)
plt.ylabel('Frequency', fontsize=18)

# Increase tick label sizes
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
import numpy as np

test_sequence_lengths = [len(sequence) for sequence in test_sequences]
min_length = min(test_sequence_lengths)
max_length = max(test_sequence_lengths)

# Plot the histogram with narrower bars to leave a gap
plt.hist(
    test_sequence_lengths,
    bins=np.arange(min_length - 0.5, max_length + 1.5, 1),  # Adjust bin edges for centering
    edgecolor='black',
    rwidth=0.8  # Leave gaps between bars
)

# Set x-axis ticks to integer values
plt.xticks(range(min_length, max_length + 1))

# Add titles and labels
plt.title('Test Sequence Lengths')
plt.xlabel('Sequence Lengths')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
